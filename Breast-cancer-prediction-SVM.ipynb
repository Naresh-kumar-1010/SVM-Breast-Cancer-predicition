# Model Training : Support Vector Machine

We will create our Support Vector Machine model using scikit-learn. SVM is used as classification techniques which first attempts to find Hyperplane, which can separate objects of different classes and then leaves maximum marigin from the nearest points of each class to the separating hyperplane.

```python
model = svm.SVC(kernel = 'linear', gamma = 'auto', C=0.3)
```

Here kernel, gamma & c are explained below:

kernel method is a class of algorithms for pattern analysis.
gamma parameter determines the distance a single data sample exerts influence or it is said to adjust the curvature of the decision boundary.
C determines penalty or how many samples are allowed to be placed in different classes.

```python
model.fit(x_train, y_train)
```

```python
y_predicted = model.predict(x_test)
```

By doing this we can find the difference between the actual data (y_test) & predicted data (y_predicted.

Further, we will evaluate the results & check the precision and f1 score.


# Results, Precision and F1 score

```python
print(classification_report(y_test, y_predicted))
```

Here,

1.Precision: Precision measures the percentage of true positives (correctly predicted positive instances) out of all predicted positives. It gives an indication of how accurate the model's positive predictions are.

Precision = True Positives / (True Positives + False Positives)

2.Recall: Recall measures the percentage of true positives out of all actual positives. It gives an indication of how well the model is able to identify positive instances.

Recall = True Positives / (True Positives + False Negatives)

3.F1-score: The F1-score is the harmonic mean of precision and recall, and is a balanced metric that takes both into account. It provides a single measure of a model's performance.

F1-score = 2 * (Precision * Recall) / (Precision + Recall)
